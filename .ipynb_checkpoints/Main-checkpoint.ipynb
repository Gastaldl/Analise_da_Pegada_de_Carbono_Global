# Análise da Pegada de Carbono Global e Indicadores de Sustentabilidade

## Introdução
  
Este projeto tem como objetivo analisar e comparar a pegada de carbono e os indicadores de sustentabilidade de diversos países, com um foco especial no Brasil. Utilizando dois conjuntos de dados principais — um com informações detalhadas sobre emissões de CO₂ fósseis por setor (como carvão, petróleo e gás) e outro com dados sobre acesso à energia, capacidade de geração de eletricidade renovável, intensidade energética e indicadores econômicos e populacionais — buscamos compreender os fatores que influenciam as emissões de carbono e identificar padrões relacionados à sustentabilidade energética.

Embora o conceito de pegada de carbono inclua tanto as emissões diretas quanto indiretas, este projeto utiliza emissões diretas de CO₂ como proxy. Isso ocorre porque as emissões de carbono representam o componente mais significativo e mensurável da pegada de carbono global, sendo amplamente usadas em análises ambientais e de sustentabilidade. 

A análise inclui uma investigação aprofundada da pegada de carbono do Brasil, comparando-a com a média global, com países das Américas e membros da OCDE. Exploramos também os impactos das políticas energéticas brasileiras, incluindo o uso de energias renováveis, e como esses esforços posicionam o Brasil em relação a grandes emissores como os Estados Unidos.

Além disso, o projeto incorpora o uso de modelos de aprendizado de máquina para realizar previsões e classificar os países em categorias de pegada de carbono com base em variáveis como PIB per capita e participação de energia renovável no consumo total. A análise utiliza métodos como regressão para entender relações quantitativas e classificação para categorizar os países. A abordagem é completa, englobando a combinação e limpeza dos dados, visualizações avançadas e uma documentação acessível, visando leitores com e sem conhecimento técnico.

## 6 - Qual é a relação entre o percentual de energia renovável no consumo total e a pegada de carbono de um país?

Neste código, vamos investigar a relação entre o percentual de energia renovável no consumo total de um país e sua pegada de carbono (emissões totais de CO₂). Para isso, usaremos um modelo de regressão linear para tentar prever as emissões totais de CO₂ com base na participação de energia renovável.

1. **Preparação dos dados**:
   - Selecionamos apenas as colunas necessárias: o percentual de energia renovável no consumo total e as emissões totais de CO₂.
   - Removemos qualquer linha com valores ausentes para garantir que o modelo funcione adequadamente.

2. **Definição das variáveis**:
   - A característica (variável independente) será o percentual de energia renovável.
   - O alvo (variável dependente) será as emissões totais de CO₂.

3. **Divisão dos dados**:
   - Dividimos o conjunto de dados em treino (70%) e teste (30%) para avaliar o desempenho do modelo.

4. **Treinamento do modelo**:
   - Inicializamos um modelo de regressão linear e o treinamos com os dados de treino.

5. **Predição e avaliação**:
   - Usamos o modelo para prever as emissões totais de CO₂ no conjunto de teste.
   - Calculamos duas métricas de desempenho:
     - **MSE (Erro Quadrático Médio)**: mede a média dos erros ao quadrado entre os valores previstos e reais.
     - **R² (Coeficiente de Determinação)**: indica a proporção da variação nos dados que o modelo consegue explicar.

```python
# Filtrar colunas relevantes e remover linhas com valores ausentes nas características e no alvo escolhidos
df_ml_limpo = df[['Renewable energy share in the total final energy consumption (%)', 'Total']].dropna()

# Definir a característica (variável independente) e o alvo (variável dependente)
X = df_ml_limpo[['Renewable energy share in the total final energy consumption (%)']]
y = df_ml_limpo['Total']  # Variável alvo: Emissões totais de CO₂

# Dividir os dados em conjuntos de treino (70%) e teste (30%)
X_treino, X_teste, y_treino, y_teste = train_test_split(X, y, test_size=0.3, random_state=42)

# Inicializar e treinar o modelo de regressão linear
modelo = LinearRegression()
modelo.fit(X_treino, y_treino)

# Fazer previsões no conjunto de teste
y_predito = modelo.predict(X_teste)

# Calcular métricas de desempenho do modelo
mse = mean_squared_error(y_teste, y_predito)
r2 = r2_score(y_teste, y_predito)

# Exibir resultados
mse, r2
```

```python
# Gráfico da relação entre energia renovável e emissões totais de CO₂
plt.figure(figsize=(10, 6))
plt.scatter(X_teste, y_teste, color='blue', alpha=0.5, label='Dados reais')
plt.plot(X_teste, y_predito, color='red', linewidth=2, label='Regressão Linear')
plt.title('Relação entre Energia Renovável e Emissões Totais de CO₂')
plt.xlabel('Percentual de Energia Renovável no Consumo Total (%)')
plt.ylabel('Emissões Totais de CO₂ (kt)')
plt.legend()
plt.grid(True)
plt.show()
```

Para responder à pergunta sobre a relação entre o percentual de energia renovável no consumo total e a pegada de carbono, utilizamos um modelo de regressão linear supervisionada, com as seguintes configurações:

- **Label/Tag (Variável Alvo)**: A variável alvo (label) é **Total CO₂ Emissions**, que representa as emissões totais de carbono de um país.
- **Feature (Variável Independente)**: A variável independente é o **Percentual de Energia Renovável no Consumo Total**.

### Resultados do Modelo

1. **MSE (Erro Quadrático Médio)**: O valor do MSE indica o erro médio quadrático das previsões em relação aos valores reais. Quanto menor o MSE, melhor a precisão do modelo. No entanto, o valor obtido aqui é relativamente alto, sugerindo que o modelo não está prevendo bem as emissões totais com base apenas na participação de energia renovável.

2. **R² (Coeficiente de Determinação)**: O valor de R² é aproximadamente 0,01, o que indica que o modelo de regressão linear explica apenas cerca de 1% da variabilidade nas emissões totais de CO₂ com base no percentual de energia renovável. Esse baixo valor de R² sugere que a participação de energia renovável no consumo total não é um forte preditor das emissões totais de CO₂ de um país isoladamente.

A análise indica que não há uma relação forte entre o percentual de energia renovável no consumo total de um país e suas emissões totais de CO₂, pelo menos no formato de uma relação linear simples. Isso pode ocorrer porque as emissões de CO₂ dependem de diversos outros fatores além da participação de energia renovável, como o uso de combustíveis fósseis, o perfil industrial, e o nível de desenvolvimento econômico.

## 9 - Com base no percentual de energia renovável no consumo total e no PIB per capita, é possível classificar os países em grupos com alta, média ou baixa pegada de carbono?

Neste código, vamos explorar a possibilidade de classificar os países em grupos de pegada de carbono — **Baixa**, **Média** ou **Alta** — usando duas variáveis: o **percentual de energia renovável no consumo total** e o **PIB per capita**. A classificação será feita com um modelo de aprendizado de máquina e avaliada por meio de uma matriz de confusão e um relatório de classificação.

1. **Tratamento dos Dados**:
   - Excluímos linhas com valores ausentes nas colunas de interesse.

2. **Criação dos Rótulos de Pegada de Carbono**:
   - Usamos o método `qcut` para dividir a coluna de emissões per capita em três categorias: Baixa, Média e Alta, com base em quantis.

3. **Preparação dos Dados para o Modelo**:
   - Selecionamos as variáveis de entrada (percentual de energia renovável e PIB per capita) e o rótulo de saída (categoria de pegada de carbono).
   - Dividimos os dados em conjuntos de treino e teste (70% e 30%, respectivamente).

4. **Treinamento e Avaliação do Modelo**:
   - Treinamos um modelo de classificação Random Forest.
   - Fazemos previsões no conjunto de teste e calculamos a precisão, o recall e o F1-score para cada classe.

5. **Análise de Resultados**:
   - Exibimos a matriz de confusão para visualizar o desempenho do modelo em cada categoria.
   - Imprimimos um relatório de classificação detalhado com as métricas de desempenho.

6. **Tabela de Comparação**:
   - Criamos uma tabela com os países no conjunto de teste, mostrando a categoria real e a categoria prevista pelo modelo, para uma análise mais específica.

```python
# Criar rótulo de pegada de carbono em categorias (Baixa, Media, Alta) com base em quantis
df['Carbon_Footprint_Category'] = pd.qcut(
    df['Per Capita'], 
    q=3, 
    labels=['Baixa', 'Media', 'Alta']
)

# Selecionar as features de entrada e a coluna de saída para o modelo
X = df[['Renewable energy share in the total final energy consumption (%)', 'gdp_per_capita']]
y = df['Carbon_Footprint_Category']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Treinar o modelo de classificação usando RandomForest
classifier = RandomForestClassifier(random_state=42)
classifier.fit(X_train, y_train)

# Fazer previsões no conjunto de teste
y_pred = classifier.predict(X_test)

classification_report_output = classification_report(y_test, y_pred, target_names=['Baixa', 'Media', 'Alta'])

# Obter a matriz de confusão
conf_matrix = confusion_matrix(y_test, y_pred, labels=['Baixa', 'Media', 'Alta'])

plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['Baixa', 'Media', 'Alta'], yticklabels=['Baixa', 'Media', 'Alta'])
plt.xlabel('Previsão')
plt.ylabel('Real')
plt.title('Matriz de Confusão para Classificação de Pegada de Carbono')
plt.show()

print("Relatório de Classificação:\n", classification_report_output)

# Adicionar as previsões ao conjunto de teste e criar uma tabela com países e categorias
X_test_with_predictions = X_test.copy()
X_test_with_predictions['Real_Category'] = y_test.values
X_test_with_predictions['Predicted_Category'] = y_pred

# Unir ao dataframe original para adicionar a coluna de país
test_results = X_test_with_predictions.join(
    df.set_index(['Renewable energy share in the total final energy consumption (%)', 'gdp_per_capita'])['Country'], 
    on=['Renewable energy share in the total final energy consumption (%)', 'gdp_per_capita']
)

# Selecionar as colunas relevantes para simplificação
test_results = test_results[['Country', 'Real_Category', 'Predicted_Category']]
test_results
```

1. **Matriz de Confusão**:
   - A matriz de confusão mostra um desempenho satisfatório do modelo, com maior precisão nas classes Baixa e Média. A classe Alta apresentou algumas previsões incorretas, o que pode ser devido à complexidade de predizer altas emissões per capita apenas com base em energia renovável e PIB.<br><br>

2. **Relatório de Classificação**:
   - As métricas indicam um bom desempenho geral, com uma **acurácia média de 84%**.
   - **F1-score** para as classes Baixa, Média e Alta mostram que o modelo é balanceado em precisão e recall, sendo mais eficiente nas classes Baixa e Média.<br><br>

3. **Tabela de Comparação com Países**:
   - A tabela final fornece uma visão detalhada das previsões por país, permitindo identificar onde o modelo acertou e onde houve erros.
  
---

**Explicação do Label/Tag Utilizado na Aprendizagem (Fit) do Modelo:**

No modelo de classificação, utilizamos a variável **"Carbon_Footprint_Category"** como o **label/tag**, que representa a categorização da pegada de carbono de cada país em três grupos: **Baixa**, **Média** e **Alta**. Essa categorização foi feita utilizando os quantis da variável **"Per Capita"** (emissões de CO₂ per capita), dividindo os dados de forma equilibrada em três partes.

Essa variável categórica serve como o objetivo de predição do modelo. Durante o treinamento, o modelo aprende a associar as features de entrada — **"Participação de energia renovável no consumo final"** e **"PIB per capita"** — às respectivas categorias de pegada de carbono. O modelo foi treinado para identificar padrões entre os indicadores econômicos e energéticos e como esses padrões influenciam a intensidade da pegada de carbono.

---

Este modelo de classificação consegue, com boa precisão, agrupar países com diferentes pegadas de carbono com base em energia renovável e PIB per capita. No entanto, os resultados indicam que alguns países podem não se encaixar perfeitamente devido à complexidade dos fatores que afetam as emissões, sugerindo que a adição de outras variáveis pode melhorar a classificação.

## 10 - Qual é a relação quantitativa entre o uso de fontes energéticas específicas (carvão, petróleo, gás) e as emissões totais de CO₂ em diferentes países?

Neste código, vamos construir e avaliar uma regressão linear para entender a relação entre o consumo de energia per capita e as emissões totais de CO₂ no Brasil. Os passos realizados incluem:

1. **Treinamento e Teste**: Dividimos os dados em 70% para treinamento e 30% para teste, garantindo que o modelo seja avaliado em dados não vistos durante o treinamento.
2. **Ajuste do Modelo**: Utilizamos o consumo de energia per capita como variável independente (X) e as emissões totais de CO₂ como variável dependente (y) para treinar o modelo.
3. **Predição e Avaliação**: Após treinar o modelo, fazemos previsões para o conjunto de teste e calculamos métricas como o erro médio quadrático (MSE) e o coeficiente de determinação (R²) para avaliar o desempenho.
4. **Visualização da Relação**: Geramos um gráfico que mostra os dados reais do teste em comparação com as previsões do modelo, incluindo a linha de regressão ajustada.

Essas etapas nos ajudarão a quantificar a influência do consumo de energia per capita nas emissões de carbono no Brasil, além de verificar se o modelo ajustado captura bem a tendência dos dados.

```python
# Filtrar os dados para o Brasil
brasil_data = df[df['Country'] == 'Brazil']

# Selecionar as variáveis de interesse
X = brasil_data[['Primary energy consumption per capita (kWh/person)']].values
y = brasil_data['Total'].values

# Dividir os dados em 70% treino e 30% teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Criar e ajustar o modelo de regressão linear
model = LinearRegression()
model.fit(X_train, y_train)

# Fazer previsões nos dados de teste
y_pred = model.predict(X_test)

# Avaliar o modelo
r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
coef = model.coef_
intercept = model.intercept_

r2, mse, coef, intercept
```

```python
# Dados reais (testes)
x_test_values = X_test.flatten()
y_test_values = y_test

# Previsões do modelo
y_pred_values = y_pred

# Criar gráfico
plt.figure(figsize=(10, 6))
plt.scatter(x_test_values, y_test_values, color='blue', label='Dados Reais', alpha=0.7)
plt.plot(x_test_values, y_pred_values, color='red', linewidth=2, label='Linha de Regressão')
plt.title('Regressão Linear: Consumo de Energia per Capita vs Emissões Totais de CO₂ (Brasil)', fontsize=14)
plt.xlabel('Consumo de Energia per Capita (kWh/pessoa)', fontsize=12)
plt.ylabel('Emissões Totais de CO₂ (kt)', fontsize=12)
plt.legend()
plt.grid(True)
plt.show()
```

### Resultado do Modelo:

   O R² de **0,96** indica que o modelo explica **96% da variabilidade** nas emissões totais de CO₂ com base no consumo de energia per capita. Isso demonstra uma forte relação linear entre essas variáveis no Brasil.

2. **Erro Médio Quadrático (MSE):**  
   O MSE de **236,63** sugere que, em média, há uma diferença relativamente pequena entre os valores reais e previstos, considerando que os valores de emissões estão na casa das centenas de milhares.

3. **Coeficiente de Regressão:**  
   O coeficiente de **0,0429** indica que, para cada aumento de 1 kWh/pessoa no consumo de energia, as emissões totais de CO₂ aumentam em aproximadamente **0,043 mil toneladas**. 

4. **Intercepto:**  
   O intercepto de **-230,16** sugere o valor estimado das emissões totais quando o consumo de energia per capita é zero. Embora esse valor não seja diretamente aplicável fisicamente, ele serve para ajustar a reta no modelo.

5. **Label/Tag**:

    No processo de aprendizado do modelo de regressão linear, a variável **"Emissões Totais de CO₂" (Total)** foi usada como a **label (ou tag)**. Essa variável representa o valor que o modelo tenta prever com base na entrada fornecida, que, neste caso, é o **"Consumo de Energia per Capita" (Primary energy consumption per capita)**.

    A escolha dessa label é fundamentada no objetivo do modelo, que é analisar como o consumo de energia influencia diretamente as emissões totais de CO₂ no Brasil. O modelo aprende a relação entre a variável independente (consumo de energia per capita) e a dependente (emissões totais de CO₂) durante o treinamento, ajustando os coeficientes da equação linear para minimizar os erros de previsão.
